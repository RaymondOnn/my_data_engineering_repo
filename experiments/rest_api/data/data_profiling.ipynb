```python
import pandas as pd  # 2.2.0
```

```python
folder_path = './data/raw/'
dep = pd.read_csv(folder_path + 'departments.csv')
dep_emp = pd.read_csv(folder_path + 'dept_emp.csv')
dep_mng = pd.read_csv(folder_path + 'dept_manager.csv')
emp = pd.read_csv(folder_path + 'employees.csv')
sal = pd.read_csv(folder_path + 'salaries.csv')
title = pd.read_csv(folder_path + 'titles.csv')
```

```python
def profile(df):
    # Unique values for each column
    display(df.head())
    for item in df.columns:
        n_values, n_rows = df[item].nunique(), df.shape[0]
        # dupes = df[item].duplicated(keep=False)
        # n_dupes = dupes.nunique()
        print(f'{item}: {n_values} values, {n_rows} rows' )
        print (df[item].unique(), "\n")
        # display(dupes.head(10))
        
for df in [dep, dep_emp, dep_mng, emp, sal, title]:
    profile(df)        
```

```python
def left_join(df1:pd.DataFrame, df2:pd.DataFrame, join_on_cols: str, new_cols: list):
    if not set(new_cols).issubset(set(df1.columns)):
        df1.merge(df2, how='left', on=join_on_cols)
    print(df1.info())
    return df1 

emp = left_join(emp, sal, 'emp_no', ['salary'])
folder_path = './data/final/'
emp.to_csv(folder_path + 'employees.csv')
```
